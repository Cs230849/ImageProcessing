# Translation


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./fuji.jpg')
print(image.shape)

translation_matrix = np.float32([[1, 0, 100], [0,1, 50]])
translated_image = cv2.warpAffine(image,
translation_matrix, (image.shape[1], image.shape[0]))
print(translated_image.shape)

translation_matrix_1 = np.float32([[1, 0, -100], [0, 1, 30]])
translated_image_1 = cv2.warpAffine(image, translation_matrix_1, (image.shape[1], image.shape[0]))
print(translated_image_1.shape)

plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(translated_image_1, cv2.COLOR_BGR2RGB))
plt.title('Translated Image_1')
plt.axis('off')
plt.show()
```

# Rotation

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotation_matrix_1 = cv2.getRotationMatrix2D(center, 90, 1.0)
rotation_matrix_2 = cv2.getRotationMatrix2D(center, 180, 2.0)


rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))
rotated_image_1 = cv2.warpAffine(image, rotation_matrix_1, (w, h))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (45 degrees)')
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(rotated_image_1, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (90 degrees)')
plt.axis('off')
plt.show()
```

# Scalling

```python
image_1 = cv2.imread('./image.jpg')
Factor = 20
Factor_1 = 0.5
scaling_matrix = cv2.getRotationMatrix2D((0, 0), 0, Factor)
scaling_matrix_1 = cv2.getRotationMatrix2D((0, 0), 0, Factor_1)

scaled_image = cv2.warpAffine(image, scaling_matrix, (w*Factor, h*Factor))
scaled_image_1 = cv2.warpAffine(image_1, scaling_matrix_1, (int(w*Factor_1), int(h*Factor_1)))

plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (20x)')
plt.axis('off')
plt.show()
print(scaled_image.shape)

plt.imshow(cv2.cvtColor(scaled_image_1, cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (0.5x)')
plt.axis('off')
plt.show()
print(scaled_image.shape)
```

# Shearing

```python
m = 0.67
m_y = 0.3

shearing_matrix = np.float32([[1, m, 0], [0, 1, 0]])
shearing_matrix_y = np.float32([[1, 0, 0], [m_y, 1, 0]])

sheared_image_y = cv2.warpAffine(image, shearing_matrix_y, (w, h))
sheared_image = cv2.warpAffine(image, shearing_matrix, (w, h))

plt.imshow(cv2.cvtColor(sheared_image, cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (x-axis, factor ' + str(m)+')')
plt.axis('off')
plt.show()
plt.show()

plt.imshow(cv2.cvtColor(sheared_image_y, cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (y-axis, factor ' + str(m_y)+')')
plt.axis('off')
plt.show()
plt.show()
```
