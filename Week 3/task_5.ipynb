```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("./horse.jpg")

if image is None:
    print("Error: could not load ./horse.jpg")
else:
    try:
        alpha_str = input('* Enter the alpha value [0.1-3.0] (default 1.5): ').strip()
        alpha = float(alpha_str) if alpha_str else 1.5
    except ValueError:
        print("Invalid alpha  using default 1.5")
        alpha = 1.5

    try:
        beta_str = input('* Enter the beta value [0-100] (default 30): ').strip()
        beta = int(beta_str) if beta_str else 30
    except ValueError:
        print("Invalid beta  using default 30")
        beta = 30

    new_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

    plt.figure(figsize=[15,5])
    plt.subplot(121)
    plt.imshow(image[..., ::-1])
    plt.title("Original")
    plt.subplot(122)
    plt.imshow(new_image[..., ::-1])
    plt.title(f"Output (={alpha}, ={beta})")
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("./horse.jpg")

if image is None:
    print("Error loading image")
else:
    new_image2 = np.zeros(image.shape, image.dtype)

    try:
        alpha_input = input('* Enter the alpha value [0.1-3.0] (press Enter for 1.5): ').strip()
        alpha = float(alpha_input) if alpha_input else 1.5
    except ValueError:
        print("Invalid alpha  using 1.5")
        alpha = 1.5

    try:
        beta_input = input('* Enter the beta value [0-100] (press Enter for 30): ').strip()
        beta = int(beta_input) if beta_input else 30
    except ValueError:
        print("Invalid beta  using 30")
        beta = 30

    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            for c in range(image.shape[2]):
                new_image2[y, x, c] = np.clip(alpha * image[y, x, c] + beta, 0, 255)

    plt.figure(figsize=[15,5])
    plt.subplot(121); plt.imshow(image[...,::-1]);     plt.title("Original")
    plt.subplot(122); plt.imshow(new_image2[...,::-1]); plt.title("Output (manual loop)")
    plt.show()
```

```python
gray_image =  cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2GRAY)
gray_new_image2 = cv2.cvtColor(new_image2, cv2.COLOR_BGR2GRAY)
```
