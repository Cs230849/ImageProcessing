```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('woman.jpeg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

divided_image_3 = cv2.divide(image_RGB, 3.0)
divided_image_0_5 = cv2.divide(image_RGB, 0.5)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(divided_image_3)
plt.title('Divided Image (Factor 3.0)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(divided_image_0_5)
plt.title('Divided Image (Factor 0.5)')
plt.axis('off')

plt.show()
```
